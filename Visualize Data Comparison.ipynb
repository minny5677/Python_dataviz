# กราฟแท่ง

### สร้างกราฟเปรียบเทียบรายได้ปี 2541 ของ กทม. และ ปริมณฑล

```python
plt.bar([1,2,3,4],greaterbangkok.iloc[:,2]) #[1,2,3,4] คือ กำหนดตำแหน่ง x ,greaterbangkok.iloc[:,2] คือ เลือกข้อมูลที่สร้างกราฟ เอาทุกแถว หลักที่ 2
```

```python
plt.bar([1,2,3,4],greaterbangkok.iloc[:,2],tick_label=greaterbangkok.iloc[:,-1]) #tick_label คือ ใส่ชื่อแต่ละแท่งโดยดึงชื่อมาจากตาราง greaterbangkok.iloc[:,-1]
plt.title(u'กราฟเปรียบเทียบรายได้ปี 2541 ของ กทม. และ ปริมณฑล',fontname='Tahoma') #fontname='Tahoma' คือ เปลี่ยนภาษา
plt.xlabel(u'จังหวัด',fontname='Tahoma') # xlabel คือ ตั้งชื่อแกน x
plt.ylabel(u'รายได้',fontname='Tahoma') # ylabel คือ ตั้งชื่อแกน y
```

### เปรียบเทียบข้อมูล 2 แท่ง ปี2541 กับ 2558

```python
plt.bar([1-(0.35/2),2-(0.35/2),3-(0.35/2),4-(0.35/2)],greaterbangkok.iloc[:,-2],width=0.35,tick_label=greaterbangkok.iloc[:,-1],color='b',label='2541',align='edge')
plt.bar([1+(0.35/2),2+(0.35/2),3+(0.35/2),4+(0.35/2)],greaterbangkok.iloc[:,2],width=0.35,tick_label=greaterbangkok.iloc[:,-1],color='r',label='2558',align='edge')
plt.legend(loc=1) #ใส่กล่องมุมบนขวา
```
