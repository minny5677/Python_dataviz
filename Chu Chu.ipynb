how many time A,T,G,C in the sequence?

```python
def char1(seq,char): #คิดก่อนว่าจะใช้ค่าอะไรบ้างแล้วสร้างตัวแปรเก็บค่านั้น
    charcount = 0
    for i in seq : #วนลูปกี่รอบ วนตามค่าอะไร
        if i == char:
            charcount += 1
    return charcount
```

```python
char1(DNA_df.iloc[1][0],'A')
```

```python
Along = 0
Tlong = 0
Glong = 0
Clong = 0
for i in range(len(DNA_df)):
    Along += char1(DNA_df.iloc[i][0],'A') #เอาฟังก์ชั่นมาใช้ .iloc เอาไว้ชี้ว่าจะเอาค่าส่วนไหนในตาราง []ตัวแรกคือแถว []ตัวหลังคือคอลัมล์
    Tlong += char1(DNA_df.iloc[i][0],'T')
    Glong += char1(DNA_df.iloc[i][0],'G')
    Clong += char1(DNA_df.iloc[i][0],'C')
print(f'A = {Along} T = {Tlong} G ={Glong} C = {Clong}')
```

```python
listp = 'ATTTATAGCGC'
```

```python
def str1(seq):
    str23 = []
    for i in range(len(seq)-4):
        str23.append(seq[i:i+5]) #sep(0:5)=(0,1,2,3,4) ตัดให้มันแสดงค่าตัวละ 5 อักษร
    return str23
```

```python
str1(DNA_df.iloc[2][0]) #เทสฟังก์ชั่นดูทุกครั้งว่าใช้ได้ไหมนะคะ
```

```python
liststr5 = []
for i in range(len(DNA_df)):
    liststr5 += str1(DNA_df.iloc[i][0])
```

```python
len(liststr5)
```

```python
unique_5char=set(liststr5) # กำจัดตัวที่ซ่ำ
```

```python
len(unique_5char)
```

```python
dict_5char = {}
for i in unique_5char:
    num = 0 #วนลูปในตัวยูนีคทั้งหมด
    for j in liststr5: #วนลูปทั้งหมดมาเทียบใส่นูนีค
        if i == j: #ถ้าซ้ำให้มันเก็บค่า
            num += 1
    dict_5char[i] = num
```

```python
dict_5char
```

```python
MaxV = -100
Maxkey = None
for i in dict_5char.keys():
    if dict_5char[i] > MaxV:
        MaxV = dict_5char[i]
        Maxkey = i
```

```python
print(f'Max 5-char-long sequence is {Maxkey} with frequency equals {MaxV}')
```
